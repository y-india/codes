```python
import pandas as pd

file_path = "all_billionaires.txt"

with open(file_path, "r", encoding="utf-8") as f:
    lines = [line.strip() for line in f if line.strip()]

data = []
i = 0

while i < len(lines):
    try:
        # skip header noise if any
        if not lines[i].isdigit():
            i += 1
            continue

        rank = int(lines[i])
        name = lines[i + 1]
        net_worth = lines[i + 2]

        arrow = lines[i + 3]
        change_amount = lines[i + 4]
        change_percent = lines[i + 5]

        industry = lines[i + 6]
        age = lines[i + 7]
        country = lines[i + 8]

        data.append([
            rank,
            name,
            net_worth,
            arrow,
            change_amount,
            change_percent,
            industry,
            age,
            country
        ])

        i += 9

    except Exception:
        i += 1

columns = [
    "Rank",
    "Name",
    "Net Worth",
    "Change Arrow",
    "Change Amount",
    "Change Percent",
    "Industry",
    "Age",
    "Country"
]

df = pd.DataFrame(data, columns=columns)

print(df.head())

```
