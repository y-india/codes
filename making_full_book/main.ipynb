```python
# File paths for all chapters/files
intro = r"making_full_book\Finding_Fertile_Land\introduction.txt"

ch1 = r"making_full_book\Finding_Fertile_Land\Chapter_1.txt"
ch2 = r"making_full_book\Finding_Fertile_Land\Chapter_2.txt"
ch3 = r"making_full_book\Finding_Fertile_Land\Chapter_3.txt"
ch4 = r"making_full_book\Finding_Fertile_Land\Chapter_4.txt"
ch5 = r"making_full_book\Finding_Fertile_Land\Chapter_5.txt"
ch6 = r"making_full_book\Finding_Fertile_Land\Chapter_6.txt"
ch7 = r"making_full_book\Finding_Fertile_Land\Chapter_7.txt"
ch8 = r"making_full_book\Finding_Fertile_Land\Chapter_8.txt"
ch9 = r"making_full_book\Finding_Fertile_Land\Chapter_9.txt"
ch10 = r"making_full_book\Finding_Fertile_Land\Chapter_10.txt"

conclusion = r"making_full_book\Finding_Fertile_Land\_Conclusion.txt"

# Store files in order
files = [
    intro,
    ch1, ch2, ch3, ch4, ch5,
    ch6, ch7, ch8, ch9, ch10,
    conclusion
]

# Output file name
output_file = "merged_book.txt"

# Separator line (60 tildes)
separator = "\n" + "~" * 60 + "\n"

# Merge files
with open(output_file, "w", encoding="utf-8") as outfile:
    for i, file in enumerate(files):
        try:
            with open(file, "r", encoding="utf-8") as infile:
                outfile.write(infile.read())

            # Add separator after every file except last
            if i < len(files) - 1:
                outfile.write(separator)

        except FileNotFoundError:
            print(f"File not found: {file}")

print(f"All files merged successfully into '{output_file}'")
```
