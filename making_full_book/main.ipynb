```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    PageBreak
)
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.lib.units import inch
```

```python
chapter_titles = [
    "Introduction",
    "Chapter 1",
    "Chapter 2",
    "Chapter 3",
    "Chapter 4",
    "Chapter 5",
    "Chapter 6",
    "Chapter 7",
    "Chapter 8",
    "Chapter 9",
    "Chapter 10",
    "Conclusion"
]
```

```python
# Input merged text file
input_txt = "merged_book.txt"

# Output PDF
output_pdf = "book_output.pdf"

# Separator used between chapters
separator = "~" * 60

# Create PDF document
doc = SimpleDocTemplate(
    output_pdf,
    rightMargin=0.8 * inch,
    leftMargin=0.8 * inch,
    topMargin=1 * inch,
    bottomMargin=1 * inch
)

styles = getSampleStyleSheet()

title_style = styles["Heading1"]
body_style = styles["BodyText"]

story = []

# Read merged text
with open(input_txt, "r", encoding="utf-8") as file:
    content = file.read()

# Split chapters using separator
chapters = content.split(separator)

# Add chapters to PDF
for i, chapter in enumerate(chapters):

    # Chapter title
    story.append(
    Paragraph(chapter_titles[i], title_style)
)

    story.append(Spacer(1, 20))

    # Convert paragraphs
    paragraphs = chapter.strip().split("\n")

    for para in paragraphs:
        para = para.strip()

        if para:
            story.append(
                Paragraph(para, body_style)
            )
            story.append(Spacer(1, 10))

    # Start next chapter on new page
    if i < len(chapters) - 1:
        story.append(PageBreak())

# Build PDF
doc.build(story)

print("PDF created successfully:", output_pdf)
```

```python
from reportlab.platypus import (
    SimpleDocTemplate,
    Paragraph,
    Spacer,
    PageBreak
)
from reportlab.lib.styles import getSampleStyleSheet, ParagraphStyle
from reportlab.lib.units import inch
from reportlab.lib.enums import TA_CENTER, TA_JUSTIFY
from reportlab.lib.pagesizes import inch as page_inch

# ======================
# SETTINGS
# ======================

input_txt = r"D:\Github_Connected\codes\making_full_book\merged_book.txt"
output_pdf = "final_book.pdf"

BOOK_TITLE = "Finding Fertile Land"

separator = "~" * 60

chapter_titles = [
    "Introduction",
    "Chapter 1",
    "Chapter 2",
    "Chapter 3",
    "Chapter 4",
    "Chapter 5",
    "Chapter 6",
    "Chapter 7",
    "Chapter 8",
    "Chapter 9",
    "Chapter 10",
    "Conclusion"
]

# 6 x 9 inch book size
PAGE_SIZE = (6 * inch, 9 * inch)

# ======================
# PDF DOCUMENT
# ======================

doc = SimpleDocTemplate(
    output_pdf,
    pagesize=PAGE_SIZE,
    leftMargin=0.75 * inch,
    rightMargin=0.75 * inch,
    topMargin=0.8 * inch,
    bottomMargin=0.8 * inch
)

styles = getSampleStyleSheet()

chapter_style = ParagraphStyle(
    "ChapterStyle",
    parent=styles["Heading1"],
    fontName="Times-Bold",
    fontSize=20,
    alignment=TA_CENTER,
    spaceAfter=30
)

body_style = ParagraphStyle(
    "BodyStyle",
    parent=styles["BodyText"],
    fontName="Times-Roman",
    fontSize=11,
    leading=18,
    alignment=TA_JUSTIFY,
    firstLineIndent=20,
    spaceAfter=10
)

story = []

# ======================
# HEADER + PAGE NUMBER
# ======================

def add_header_footer(canvas, doc):
    canvas.saveState()

    # Header
    canvas.setFont("Times-Roman", 9)
    canvas.drawCentredString(
        PAGE_SIZE[0] / 2,
        PAGE_SIZE[1] - 30,
        BOOK_TITLE
    )

    # Footer Page Number
    canvas.drawCentredString(
        PAGE_SIZE[0] / 2,
        20,
        str(canvas.getPageNumber())
    )

    canvas.restoreState()

# ======================
# READ TEXT
# ======================

with open(input_txt, "r", encoding="utf-8") as file:
    content = file.read()

chapters = content.split(separator)

# ======================
# BUILD CONTENT
# ======================

for i, chapter in enumerate(chapters):

    title = (
        chapter_titles[i]
        if i < len(chapter_titles)
        else f"Chapter {i}"
    )

    story.append(
        Paragraph(title, chapter_style)
    )

    story.append(
        Spacer(1, 20)
    )

    paragraphs = chapter.strip().split("\n")

    for para in paragraphs:

        para = para.strip()

        if para:
            story.append(
                Paragraph(para, body_style)
            )

    if i < len(chapters) - 1:
        story.append(
            PageBreak()
        )

# ======================
# GENERATE PDF
# ======================

doc.build(
    story,
    onFirstPage=add_header_footer,
    onLaterPages=add_header_footer
)

print("Book-style PDF created successfully.")
print("Saved as:", output_pdf)
```
