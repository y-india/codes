```python
with open("pytest-basic_to_advance.txt", "r", encoding="utf-8") as f:
    content = f.read()

# Find all positions of target
indices = []
start = 0

while True:
    pos = content.find(target, start)
    if pos == -1:
        break
    indices.append(pos)
    start = pos + 1

chunks = []

for i in range(len(indices)):
    start_idx = indices[i]
    end_idx = indices[i + 1] if i + 1 < len(indices) else len(content)
    chunk = content[start_idx:end_idx].strip()
    chunks.append(chunk)

# write files
for i, chunk in enumerate(chunks, start=1):
    filename = f"ch{i}_from_file.txt"
    with open(filename, "w", encoding="utf-8") as f:
        f.write(chunk)

print("Total chunks created:", len(chunks))
```

```python
import os

folder_path = r"C:\Users\User\OneDrive\Desktop\making_pytest_notes"
target = "the next tutorial"

for filename in os.listdir(folder_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(folder_path, filename)

        with open(file_path, "r", encoding="utf-8") as f:
            content = f.read().strip()

        if content.endswith(target):
            print(f"{filename}: YES")
        else:
            print(f"{filename}: NO")
```
